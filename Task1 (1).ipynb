```python
df['Votes'] = scaler.fit_transform(df[['Votes']])
```

```python
df = pd.get_dummies(df, columns=['Has Table booking', 'Has Online delivery', 'Is delivering now'], drop_first=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('Aggregate rating',axis=1)
y=df['Aggregate rating']
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
model_lnr = LinearRegression()  
model_lnr.fit(X_train, y_train)
predictions_lnr = model_lnr.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor()
model_rf.fit(X_train, y_train)
predictions_rf = model_rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,root_mean_squared_error
```

```python
print("Linear Regression model metrics")
print("RÂ² Score:", r2_score(y_test, predictions_lnr))
print("MAE:", mean_absolute_error(y_test, predictions_lnr))
print("MSE:", mean_squared_error(y_test, predictions_lnr))
print("RMSE:", root_mean_squared_error(y_test, predictions_lnr))
```

```python
print("Random Forest Regressor model metrics")
print("RÂ² Score:", r2_score(y_test, predictions_rf))
print("MAE:", mean_absolute_error(y_test, predictions_rf))
print("MSE:", mean_squared_error(y_test, predictions_rf))
print("RMSE:", root_mean_squared_error(y_test, predictions_rf))
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=predictions_lnr)
plt.xlabel("Actual Ratings")
plt.ylabel("Predicted Ratings")
plt.title("Actual vs Predicted Ratings (Linear regression)")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--') # Perfect prediction line
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=predictions_rf)
plt.xlabel("Actual Ratings")
plt.ylabel("Predicted Ratings")
plt.title("Actual vs Predicted Ratings (Random Forest)")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--') # Perfect prediction line
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test.values[:100], label='Actual', marker='o')
plt.plot(predictions_lnr[:100], label='Predicted', marker='x')
plt.title("Actual vs Predicted Ratings (First 100 Samples)")
plt.xlabel("Sample Index")
plt.ylabel("Aggregate Rating")
plt.legend()
plt.grid(True)
plt.show()
```

In this task, the objective was to predict the aggregate rating of restaurants based on various features available in the dataset.
I began by loading the dataset and exploring it using functions like head, info, and describe to understand its structure.

I then checked for missing values and found that the â€˜Cuisinesâ€™ column had a few null entries, which I replaced with â€˜Unknownâ€™.

Several columns like â€˜Restaurant Nameâ€™, â€˜Addressâ€™, â€˜Localityâ€™, â€˜Locality Verboseâ€™, â€˜Switch to order menuâ€™, â€˜Rating textâ€™, and â€˜Currencyâ€™
were dropped as they either had high cardinality, were repetitive, or did not contribute meaningfully to prediction. 

For categorical columns with â€˜Yesâ€™ or â€˜Noâ€™ values such as â€˜Has Table bookingâ€™, â€˜Has Online deliveryâ€™, and â€˜Is delivering nowâ€™,
I applied one-hot encoding.

Next, I explored the relationships between features and the target variable using correlation analysis for numerical columns and visualizations 
like bar plots for categorical ones.

I scaled numerical features like 'Average Cost for two' and 'Votes' to bring them onto a comparable range. 
After preprocessing, I selected relevant features and split the data into training and testing sets.

I trained two models: Linear Regression and Random Forest Regressor.
The evaluation metrics revealed that Random Forest outperformed Linear Regression, with a much higher RÂ² score and significantly lower error metrics like
MAE, MSE, and RMSE. 

To better understand model performance, I visualized actual vs predicted ratings using line plots and scatter plots.
Overall, the Random Forest model was a strong predictor of restaurant ratings
